```python
import matplotlib.pyplot as plt
import random
piezas=5000000
exito=0 #y es menor que x y embonan
for i in range(0,piezas):
  x=random.normalvariate(19,0.0016)
  y=random.normalvariate(17,0.0011)
 
  if y<x:
    exito=exito+1
probabilidad=exito/piezas
print("Embonaron",exito,"de",piezas,"piezas, la probabilidad de que NO embonen es:",1-probabilidad)
```

```python
import matplotlib.pyplot as plt

plt.hist([random.normalvariate(19, 0.0017) for i in range(piezas)], bins = 100, color="red") #pieza x
plt.title('Distribución Normal para X')
plt.grid()
plt.show()


plt.hist([random.normalvariate(17, 0.0011) for i in range(piezas)], bins = 100) #pieza y
plt.title('Distribución Normal para Y')
plt.grid()
plt.show()
```
